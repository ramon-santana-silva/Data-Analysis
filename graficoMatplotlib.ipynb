```python
#importação das bibliotecas
import matplotlib.pyplot as plt

import seaborn as sns

import plotly.express as px

import plotly.graph_objects as go

import pandas as pd

import numpy as np
```

```python
#grafico com MATPLOTLIB 
#criar uma matriz com a função linspace da numpy
""""neste exemplo a matriz foi dividida em 100 indo do intervalor de 0.0 a 2.0"""
t=np.linspace(0.0,2.0,100)
#display(t)
plt.figure(figsize=(10,10))


plt.plot(t, linewidth=5.0, label="Matriz T")

plt.legend()
plt.title("grafico da matriz")

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5)
```
